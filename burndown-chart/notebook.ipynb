```python
import altair as alt
import pandas as pd
```

```python
actual = [
    { "sprint": 1, "done": 0 },
    { "sprint": 2, "done": 0 },
    { "sprint": 3, "done": 0 },
    { "sprint": 4, "done": 0 },
    { "sprint": 5, "done": 1 },
    { "sprint": 6, "done": 1 },
]

planned = [
    { "sprint": 1, "done": 2 },
    { "sprint": 2, "done": 3 },
    { "sprint": 3, "done": 5 },
    { "sprint": 4, "done": 6 },
    { "sprint": 5, "done": 8 },
    { "sprint": 6, "done": 12 },
]

actual_df = pd.DataFrame(actual)
planned_df = pd.DataFrame(planned)

actual_df["status"] = "actual"
planned_df["status"] = "planned"

data = pd.concat([actual_df, planned_df])

burndown_chart = alt.Chart(data) \
.mark_line(point=alt.OverlayMarkDef(size=150, filled=True)) \
.encode(
    x=alt.X("sprint:O", axis=alt.Axis(labelAngle=0)).title("sprints"),
    y=alt.Y("done:N").title("user stories"),
    color=alt.Color("status:N"),
    tooltip=["sprint", "done", "status"],
) \
.properties(
    width=400,
    height=500,
) \
.interactive()

labels = alt.Chart(data).mark_text(align="center", dy=-15, fontSize=12) \
.encode(
    x="sprint:N",
    y="done:N",
    text="done:N"
)

burndown_chart = burndown_chart + labels

burndown_chart
```
